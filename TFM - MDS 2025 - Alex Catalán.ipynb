### <b>1. INTRODUCCI√ìN</b>

La **retenci√≥n estudiantil** en las Instituciones de Educaci√≥n Superior T√©cnico Profesional en Chile (**IP** y **CFT**) es un desaf√≠o cr√≠tico que impacta la **calidad educativa**, la **eficiencia institucional** y la **equidad en el acceso a la educaci√≥n**. La tasa de retenci√≥n, que mide la continuidad de los estudiantes en su trayectoria acad√©mica, var√≠a significativamente seg√∫n factores individuales, acad√©micos, econ√≥micos e institucionales. A pesar de los esfuerzos por mejorar la permanencia estudiantil, los niveles de deserci√≥n siguen siendo elevados, afectando tanto a las instituciones como a la empleabilidad futura de los estudiantes.

Seg√∫n datos del Servicio de Informaci√≥n de Educaci√≥n Superior (SIES) y la Comisi√≥n Nacional de Acreditaci√≥n (CNA), las tasas de retenci√≥n en los IP y CFT muestran desaf√≠os persistentes, con niveles de abandono que superan el 30% en algunos programas durante el primer a√±o. Entre las principales causas identificadas se encuentran:
*	**Factores individuales**: Falta de motivaci√≥n, baja autoconfianza y dificultades en la adaptaci√≥n al entorno educativo.
*	**Factores acad√©micos**: Deficiencias en formaci√≥n previa, m√©todos de ense√±anza poco efectivos y cargas acad√©micas excesivas.
*	**Factores econ√≥micos**: Falta de financiamiento, dificultades para compatibilizar estudios y trabajo.
*	**Factores institucionales**: Carencia de programas de apoyo, infraestructura deficiente y escasa personalizaci√≥n en el seguimiento del estudiante.
*	**Factores socioculturales**: Falta de redes de apoyo, diferencias en expectativas y nivel socioecon√≥mico.

Dado este panorama, surge la necesidad de contar con herramientas avanzadas que permitan **anticipar el riesgo de abandono estudiantil** para que las instituciones puedan implementar **estrategias de retenci√≥n efectivas**.

```python
def procesar_df_instituciones_sies(tabla, df_x_per, cols):
    match tabla:
        case 't_info_gral_cft_ip_sies':
            # Extracci√≥n de las columnas
            df_temp = df_x_per.iloc[:, :6]

            # Eliminaci√≥n de la columna 'Unnamed: 3' asociada a la 'Autonom√≠a' que se incluir√° en el dataframe t_info_acred_cft_ip_sies.
            df_temp = df_temp.drop(df_temp.columns[3], axis=1)

            # Renombrar las columnas del nuevo dataframe
            df_temp.columns = cols

            # ELiminar la primera fila del nuevo dataframe que contiene los nombres de las columnas 
            # (esto ocurri√≥ por defecto al importar los datos de la planilla excel al dataframe)
            df_temp = df_temp.iloc[1:]

            # Eliminar las filas de la columna 'cod_ies' que contienen NaN (son las √∫ltimas filas que se agregaron por defecto al importar los datos de la planilla excel)
            df_temp = df_temp.dropna(subset=['cod_ies'])

            # Eliminar las filas que no est√©n asociadas a CFT o IP; en otras palabras, eliminar las filas que contengan registros de Universidades.
            df_temp = df_temp[~df_temp['tipo_inst'].str.contains('Universidad', case=False, na=False)]

            # Verificar si hay duplicados en el √≠ndice
            duplicados = df_temp['cod_ies'].duplicated(keep=False)

            if duplicados.any():
                print("> ¬°Hay duplicados en el √≠ndice!, se debe revisar el dataframe.")
                print(f'> Duplicados: {df_temp[duplicados]}')
            else:
                print("> ¬°No hay duplicados en el √≠ndice!")
                print(f'> Dataframe: df_temp. Dimensi√≥n: {df_temp.shape}')
                df_temp.head(3)
        case 't_info_sii_cft_ip_sies':
            # Extracci√≥n de las columnas
            df_temp = df_x_per.iloc[:, [0, 1, 12, 11]]
            
            # Renombrar las columnas del nuevo dataframe
            df_temp.columns = cols
            
            # ELiminar la primera fila del nuevo dataframe que contiene los nombres de las columnas
            df_temp = df_temp.iloc[1:]

            # Eliminar las filas de la columna 'cod_ies' que contienen NaN (son las √∫ltimas filas que se agregaron por defecto al importar los datos de la planilla excel)
            df_temp = df_temp.dropna(subset=['cod_ies'])

            # Eliminar las filas que no est√©n asociadas a CFT o IP; en otras palabras, eliminar las filas que contengan registros de Universidades.
            df_temp = df_temp[~df_temp['tipo_inst'].str.contains('Universidad', case=False, na=False)]

            # Verificar si hay duplicados en el √≠ndice
            duplicados = df_temp['cod_ies'].duplicated(keep=False)

            if duplicados.any():
                print("> ¬°Hay duplicados en el √≠ndice!, se debe revisar el dataframe.")
                print(f'> Duplicados: {df_temp[duplicados]}')
            else:
                print("> ¬°No hay duplicados en el √≠ndice!")
                print(f'> Dataframe: df_temp. Dimensi√≥n: {df_temp.shape}')
                df_temp.head(3)
        case 't_info_acred_cft_ip_sies':
            # Extracci√≥n de las columnas. Se repite la columna 'cod_ies' para que coincida con la cantidad de columnas de la lista 'cols'.
            df_temp = df_x_per.iloc[:, [0, 0, 1, 3, 6, 7, 8, 9, 10]]
            
            # Renombrar las columnas del nuevo dataframe
            df_temp.columns = cols
            
            # ELiminar la primera fila del nuevo dataframe que contiene los nombres de las columnas
            df_temp = df_temp.iloc[1:]

            # Eliminar las filas de la columna 'cod_ies' que contienen NaN (son las √∫ltimas filas que se agregaron por defecto al importar los datos de la planilla excel)
            df_temp = df_temp.dropna(subset=['cod_ies'])

            # Eliminar las filas que no est√©n asociadas a CFT o IP; en otras palabras, eliminar las filas que contengan registros de Universidades.
            df_temp = df_temp[~df_temp['tipo_inst'].str.contains('Universidad', case=False, na=False)]

            # Verificar si hay duplicados en el √≠ndice
            duplicados = df_temp['cod_ies'].duplicated(keep=False)

            if duplicados.any():
                print("> ¬°Hay duplicados en el √≠ndice!, se debe revisar el dataframe.")
                print(f'> Duplicados: {df_temp[duplicados]}')
            else:
                print("> ¬°No hay duplicados en el √≠ndice!")
                print(f'> Dataframe: df_temp. Dimensi√≥n: {df_temp.shape}')
                df_temp.head(3)
        case _:
            print('La tabla ingresada no est√° en la lista!')
    
    return df_temp
```

‚ö° **[asigna_anio_periodo]**: Funci√≥n que permite asignar el a√±o mayor del periodo a trav√©s de la extracci√≥n de un valor en una cadena de texto.

```python
def asigna_anio_periodo(columna, periodo):
    # anio = 0
    match columna:
        case 'anio_arancel_anual_carr':            
            anio = int(periodo.split('-')[1]) # Extraer el a√±o mayor del periodo
        case 'anio_mat_tot_carr':
            anio = int(periodo.split('-')[0]) # Extraer el a√±o menor del periodo
        case 'anio_mat_1er_anio_carr':
            anio = int(periodo.split('-')[0]) # Extraer el a√±o menor del periodo
        case 'anio_tit_carr':
            anio = int(periodo.split('-')[0]) - 1 # Extraer el a√±o menor del periodo menos 1 a√±o
    return anio
```

```python
# Imputaci√≥n de valores nulos de forma segura y sin errores en columnas vac√≠as
for name, df in dataframes.items():
    df_cleaned = df.copy()  # Crear una copia expl√≠cita del DataFrame
    
    # Imputaci√≥n de valores num√©ricos con la mediana solo si la columna tiene valores no nulos
    numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        if df_cleaned[col].notnull().sum() > 0:  # Solo aplicar si hay valores no nulos
            df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

    # Imputaci√≥n de valores categ√≥ricos con la moda solo si la columna tiene valores no nulos
    categorical_cols = df_cleaned.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df_cleaned[col].notnull().sum() > 0:  # Solo aplicar si hay valores no nulos
            df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])

    # Guardar el dataframe procesado en el diccionario
    dataframes[name] = df_cleaned
```

```python
# Aplicar transformaci√≥n en el dataset "t_ret" con validaci√≥n de valores negativos
if "ret" in dataframes:
    df = dataframes["ret"].copy()  # Crear copia para evitar modificar el original

    # Definir las columnas objetivo a limpiar
    target_columns = ["tasa_ret_1er_anio", "tasa_ret_2do_anio"]

    for col in target_columns:
        if col in df.columns:
            # Convertir todos los valores a string, eliminar espacios en blanco y forzar min√∫sculas
            df[col] = df[col].astype(str).str.strip().str.lower()

            # Reemplazar cualquier variaci√≥n de 's/i' con NaN y corregir separadores decimales
            df[col] = df[col].replace(r"[^0-9\.\,]", "", regex=True)  # Mantener solo n√∫meros, punto y coma
            df[col] = df[col].replace({"s/i": np.nan, ",": "."}, regex=True)

            # Verificar si a√∫n existen valores no convertibles antes de aplicar pd.to_numeric()
            invalid_values = df[~df[col].str.replace(r"[\d\.]", "", regex=True).eq("")]

            if not invalid_values.empty:
                print(f"‚ö†Ô∏è Advertencia: Se detectaron valores que no pueden convertirse a num√©rico en {col}.")
                print(invalid_values)

            # Convertir a float de manera segura
            df[col] = pd.to_numeric(df[col], errors="coerce")

            # Validaci√≥n: Eliminar valores negativos reemplaz√°ndolos por NaN
            df[col] = df[col].apply(lambda x: x if (x >= 0 and x <= 1) else np.nan)

            # Mostrar advertencia si existen valores fuera de rango
            invalid_out_of_range = df[col].isna().sum()
            if invalid_out_of_range > 0:
                print(f"‚ö†Ô∏è Se detectaron {invalid_out_of_range} valores fuera del rango [0,1] en {col}. Fueron reemplazados por NaN.")

            # Verificar cu√°ntos valores NaN quedaron tras la conversi√≥n
            print(f"Valores NaN en {col} despu√©s de la conversi√≥n: {df[col].isnull().sum()}")

            # Aplicar imputaci√≥n de valores NaN
            if df[col].isnull().sum() > 0:
                df[col] = df.groupby(["id_carr_gen", "tipo_inst"])[col].transform(lambda x: x.fillna(x.mean()))
                df[col] = df.groupby("tipo_inst")[col].transform(lambda x: x.fillna(x.mean()))
                df[col] = df[col].fillna(df[col].mean())  # Imputaci√≥n global si a√∫n hay NaN

            # Verificar nuevamente los valores NaN despu√©s de la imputaci√≥n
            print(f"Valores NaN en {col} despu√©s de la imputaci√≥n: {df[col].isnull().sum()}")

    # Actualizar el dataset en el diccionario
    dataframes["ret"] = df

    print("‚úÖ Proceso de conversi√≥n, validaci√≥n de rango e imputaci√≥n completado correctamente.")

else:
    print("‚ùå ERROR: El dataset 't_ret' no se encuentra en el diccionario.")
```

5. <b>Detecci√≥n y eliminaci√≥n de outliers.</b><br>
<blockquote>
    Se identifican valores at√≠picos usando el rango intercuart√≠lico (IQR) y se eliminan los datos que est√°n fuera de los l√≠mites.
</blockquote>

```python
for name, df in dataframes.items():
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    dataframes[name] = df  # Guardar el dataframe actualizado
```

<b>E) Selecci√≥n final de atributos para el modelo.</b>

Basado en el an√°lisis exploratorio de datos, se seleccionaron las siguientes variables para alimentar el modelo:

üìå **Variables Objetivo (Target)**
* **<code>tasa_ret_1er_anio</code>** (Tasa de retenci√≥n en primer a√±o)
* **<code>tasa_ret_2do_anio</code>** (Tasa de retenci√≥n en segundo a√±o)

üìå **Variables Predictoras (Features)**
* **<code>id_carr_gen</code>** ‚Üí C√≥digo de carrera gen√©rica (posible impacto en la retenci√≥n seg√∫n correlaci√≥n)
* **<code>coho_mat</code>** ‚Üí Cohorte de matr√≠cula (podr√≠a influir en la retenci√≥n seg√∫n dispersi√≥n)
* **<code>area_carr</code>** ‚Üí √Årea de la carrera (diferencias en retenci√≥n observadas en los boxplots)
* **<code>tipo_inst</code>** ‚Üí Tipo de instituci√≥n (posible factor diferenciador seg√∫n retenci√≥n)

```python
# Copiar el dataset "ret" del diccionario para no modificar el original
df = dataframes["ret"].copy()

# üìå 1Ô∏è‚É£ Variables seleccionadas seg√∫n an√°lisis
target_columns = ["tasa_ret_1er_anio", "tasa_ret_2do_anio"]
numerical_features = ["id_carr_gen", "coho_mat"]
categorical_features = ["area_carr", "tipo_inst"]

# üìå 2Ô∏è‚É£ Validaci√≥n de valores negativos en tasas de retenci√≥n
for col in target_columns:
    df = df[df[col] >= 0]  # Filtra valores negativos

# üìå 3Ô∏è‚É£ Normalizaci√≥n de Variables Num√©ricas
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# üìå 4Ô∏è‚É£ Codificaci√≥n de Variables Categ√≥ricas
encoder = OneHotEncoder(drop="first", sparse_output=False)  # Evita colinealidad
encoded_cats = encoder.fit_transform(df[categorical_features])
encoded_cats_df = pd.DataFrame(
    encoded_cats,
    columns=encoder.get_feature_names_out(categorical_features),
    index=df.index,
)

# üìå 5Ô∏è‚É£ Concatenar todas las variables preparadas
df_final = pd.concat([df[target_columns], df[numerical_features], encoded_cats_df], axis=1)

# üìå 6Ô∏è‚É£ Guardar el dataset procesado en el diccionario
dataframes["ret_processed"] = df_final

print("‚úÖ Transformaci√≥n de atributos completada correctamente.")
```

<b>F) Desarrollo del Modelo de Machine Learning.</b>

En este proceso, se desarroll√≥ un modelo de Machine Learning utilizando Regresi√≥n Lineal para predecir la retenci√≥n de estudiantes en Instituciones de Educaci√≥n T√©cnico-Profesional en Chile (CFT e IP). A continuaci√≥n, se explica detalladamente cada paso, desde la carga de datos hasta la evaluaci√≥n del modelo, para comprender qu√© significa cada m√©trica obtenida y c√≥mo se podr√≠an mejorar los resultados.

<blockquote>
    üìå Paso 1: Cargar el dataset preprocesado y verificar la estructura.
</blockquote>

* Antes de entrenar cualquier modelo, es fundamental asegurarse de que los datos han sido correctamente procesados y est√°n listos para su uso.
* Si los datos contienen valores nulos, errores o si alguna variable importante falta, el modelo no podr√° entrenarse correctamente o arrojar√° resultados err√≥neos.

```python
# Cargar el dataset procesado desde el diccionario
if "ret_processed" in dataframes:
    df_model = dataframes["ret_processed"].copy()
    print("‚úÖ Dataset 'ret_processed' cargado correctamente.")
else:
    raise ValueError("‚ùå ERROR: El dataset 'ret_processed' no est√° en el diccionario. Verifica el preprocesamiento.")
    
# Validar estructura del dataframe
if df_model.empty:
    raise ValueError("‚ùå ERROR: El dataset est√° vac√≠o despu√©s del preprocesamiento.")
else:
    print(f"üìä Dimensiones del dataset: {df_model.shape}")
    print(f"üìã Columnas disponibles: {df_model.columns.tolist()}")
```

<blockquote>
    üìå Paso 2: Definir variables objetivo y predictoras
</blockquote>

En este paso, se seleccionaron las variables que se utilizar√°n en el modelo de Machine Learning. La correcta elecci√≥n de variables es fundamental para mejorar la precisi√≥n del modelo y evitar ruido innecesario.

* **Variables Objetivo (Target)**: Son aquellas que quieren predecir. En este caso, interesa conocer la tasa de retenci√≥n de los estudiantes en el primer a√±o y en el segundo a√±o.
* **Variables Predictoras (Features)**: Son los factores que pueden influir en la retenci√≥n. Estas variables se usar√°n para entrenar el modelo y buscar patrones.


```python
# Definir variables
target_columns = ["tasa_ret_1er_anio", "tasa_ret_2do_anio"]
features = [col for col in df_model.columns if col not in target_columns]

# Verificar si las variables existen en el dataframe
missing_cols = [col for col in target_columns if col not in df_model.columns]
if missing_cols:
    raise ValueError(f"‚ùå ERROR: Las siguientes variables objetivo no se encuentran en el dataset: {missing_cols}")
else:
    print("‚úÖ Variables objetivo encontradas correctamente.")

# Separar X (features) e Y (targets)
X = df_model[features]
Y = df_model[target_columns]

print(f"üìä Variables predictoras seleccionadas: {len(features)}")
```

<blockquote>
    üìå Paso 3: Divisi√≥n del Dataset en Conjunto de Entrenamiento y Prueba
</blockquote>

El dataset fue dividido en dos partes:

* 80% para entrenamiento (X_train, Y_train)
* 20% para prueba (X_test, Y_test)

Observaciones:

* Se divide el dataset para asegurarse que el modelo pueda hacer predicciones sobre datos nuevos que nunca ha visto antes. Esto ayuda a medir su capacidad de generalizaci√≥n.

* Si no se divide el dataset, el modelo memorizar√≠a los datos en lugar de aprender patrones generales, lo que provocar√≠a un sobreajuste y una baja precisi√≥n en datos nuevos.

```python
# Divisi√≥n en 80% entrenamiento y 20% prueba
try:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    print(f"‚úÖ Divisi√≥n de datos completada con √©xito.")
    print(f"üìä Conjunto de Entrenamiento: {X_train.shape[0]} registros")
    print(f"üìä Conjunto de Prueba: {X_test.shape[0]} registros")
except Exception as e:
    raise ValueError(f"‚ùå ERROR en la divisi√≥n de datos: {e}")
```

<blockquote>
    üìå Paso 4: Entrenamiento del modelo de Machine Learning
</blockquote>

Para entrenar el modelo se utiliz√≥ Regresi√≥n Lineal, que es un algoritmo que busca la mejor relaci√≥n matem√°tica entre las variables predictoras y la variable objetivo. De acuerdo a esto:
* Es f√°cil de interpretar y visualizar.
* Permite entender qu√© factores tienen m√°s peso en la retenci√≥n.
* Sirve como modelo base para comparar con otros modelos m√°s complejos como:
    - Random Forest: Captura relaciones no lineales y ofrece mejores predicciones.
    - XGBoost: M√°s potente para detectar patrones complejos en los datos.
    - Redes Neuronales: Recomendadas si el dataset tuviera una cantidad mucho mayor de datos.

Sin embargo, para este caso inicial, la Regresi√≥n Lineal permiti√≥ establecer una base y evaluar la viabilidad del modelo.


```python
from sklearn.linear_model import LinearRegression
from sklearn.exceptions import NotFittedError

# Inicializar el modelo
modelo = LinearRegression()

# Entrenar el modelo y verificar que se haya completado correctamente
try:
    modelo.fit(X_train, Y_train)
    print("‚úÖ Modelo entrenado correctamente.")
except Exception as e:
    raise ValueError(f"‚ùå ERROR en el entrenamiento del modelo: {e}")
```

<blockquote>
    üìå Paso 5: Predicci√≥n y evaluaci√≥n del modelo
</blockquote>

Despu√©s de entrenar el modelo, se realizan predicciones sobre el conjunto de prueba (X_test) y comparan los valores predichos con los valores reales (Y_test).

```python
# Intentar hacer predicciones
try:
    Y_pred = modelo.predict(X_test)
except NotFittedError:
    raise ValueError("‚ùå ERROR: El modelo no ha sido entrenado correctamente.")
except Exception as e:
    raise ValueError(f"‚ùå ERROR al hacer predicciones: {e}")

# Evaluaci√≥n de m√©tricas
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"üìä Evaluaci√≥n del Modelo:")
print(f"üîπ MAE (Error Absoluto Medio): {mae:.4f}")
print(f"üîπ MSE (Error Cuadr√°tico Medio): {mse:.4f}")
print(f"üîπ R¬≤ Score (Coeficiente de Determinaci√≥n): {r2:.4f}")
```

##### 3.3.4 Resultados.

**MAE = 0.0931**
* Descripci√≥n: El MAE mide el error promedio absoluto entre las predicciones del modelo y los valores reales. 
* Resultados: En este caso, en promedio, la predicci√≥n del modelo se equivoca en 0.0931 unidades de retenci√≥n.
* Valores ideales: Un MAE cercano a 0 indica un modelo m√°s preciso.

**MSE = 0.0176**
* Descripci√≥n: El MSE calcula el error cuadr√°tico medio, penalizando los errores grandes. 
* Resultados: Este valor indica que, en promedio, el error al predecir la tasa de retenci√≥n es de 0.0176 unidades.
* Valores ideales: Cuanto m√°s cercano a 0, mejor ser√° el modelo.

**R¬≤ Score = 0.2170**
* Descripci√≥n: El R¬≤ Score mide qu√© porcentaje de la variabilidad en la tasa de retenci√≥n es explicada por el modelo.
* Resultados: En este caso, un R¬≤ de 0.2170 significa que el modelo solo explica el 21.7% de la variabilidad en la retenci√≥n. Esto indica que hay otros factores no considerados en los datos que podr√≠an influir en la retenci√≥n.
* Rango de valores:
    * Si R¬≤ es 1, significa que el modelo explica toda la variabilidad.
    * Si R¬≤ es 0, significa que el modelo no explica nada de la variabilidad.
